## 📦 Section 1 — Variables & Data Types
### ⏱️ Target: 6 minutes

**IT Analogy:**
> A variable is a **named memory register**. You assign a value, give it a label, and use that label later.  
> Unlike C or Java — no need to declare the type first. Python figures it out automatically.

### The 5 core data types you'll use every day:

````python
print("ok")
````

````python
# ── THE 5 CORE DATA TYPES ─────────────────────────────────────────────────

# 1. String — text, wrapped in single or double quotes
server_name = "prod-db-01"
environment = 'production'

# 2. Integer — whole numbers, no quotes
port_number   = 5432
cpu_cores     = 16

# 3. Float — decimal numbers
cpu_usage     = 87.4
disk_used_gb  = 234.7

# 4. Boolean — True or False (capital T and F)
is_online     = True
backup_done   = False

# 5. None — represents 'no value' / null
last_error    = None

# Print values with type information
print("=" * 45)
print("Variable         Value          Type")
print("=" * 45)
variables = [
    ("server_name",  server_name),
    ("port_number",  port_number),
    ("cpu_usage",    cpu_usage),
    ("is_online",    is_online),
    ("last_error",   last_error),
]
for name, val in variables:
    print(f"{name:<16} {str(val):<14} {type(val).__name__}")
print("=" * 45)
````

````python
# ── ARITHMETIC & VARIABLE REASSIGNMENT ───────────────────────────────────
# Variables can be updated. Python evaluates the right side first.

total_disk_gb  = 500
used_disk_gb   = 234.7
free_disk_gb   = total_disk_gb - used_disk_gb
usage_percent  = (used_disk_gb / total_disk_gb) * 100

print(f"Total disk   : {total_disk_gb} GB")
print(f"Used         : {used_disk_gb} GB")
print(f"Free         : {free_disk_gb} GB")
print(f"Usage        : {usage_percent:.1f}%")   # :.1f = 1 decimal place

# Common arithmetic operators
print("\n--- Operators ---")
print(f"10 + 3  = {10 + 3}")
print(f"10 - 3  = {10 - 3}")
print(f"10 * 3  = {10 * 3}")
print(f"10 / 3  = {10 / 3:.4f}")   # always returns float
print(f"10 // 3 = {10 // 3}")       # floor division (integer result)
print(f"10 % 3  = {10 % 3}")        # modulo (remainder)
print(f"10 ** 3 = {10 ** 3}")       # power
````

````python
# ── TYPE CONVERSION ───────────────────────────────────────────────────────
# IT analogy: like casting types in any typed language, or parsing a string from a config file

# From a config file, everything comes in as a string
config_port    = "8080"         # string from config
config_timeout = "30.5"         # string from config
config_debug   = "True"         # string from config

# Convert to the right types
port    = int(config_port)       # str → int
timeout = float(config_timeout)  # str → float
debug   = config_debug == "True" # str → bool (comparison)

print(f"port    = {port}   (type: {type(port).__name__})")
print(f"timeout = {timeout} (type: {type(timeout).__name__})")
print(f"debug   = {debug}  (type: {type(debug).__name__})")

# Convert number back to string (e.g. for building a URL)
base_url  = "http://api.company.com:"
full_url  = base_url + str(port)
print(f"\nFull URL: {full_url}")
````

## ✏️ Section 2 — Strings & f-strings
### ⏱️ Target: 5 minutes

**IT Analogy:**
> Strings in Python work like **template strings** in shell scripts or log formatters.  
> f-strings are Python's version of `printf` — embed variables directly inside text.

````python
# ── STRING BASICS ─────────────────────────────────────────────────────────

hostname = "prod-web-01"

# Common string operations
print(hostname.upper())          # PROD-WEB-01
print(hostname.lower())          # prod-web-01
print(hostname.replace("-","_")) # prod_web_01
print(len(hostname))             # 11 (character count)
print(hostname.startswith("prod"))  # True
print(hostname.endswith("01"))      # True
print("web" in hostname)            # True — substring check

# Splitting — very useful for parsing log lines
log_line = "2024-01-15 08:11:02 ERROR Database query timeout"
parts    = log_line.split(" ", 3)   # split on space, max 3 splits
print("\nParsed log line:")
print(f"  Date   : {parts[0]}")
print(f"  Time   : {parts[1]}")
print(f"  Level  : {parts[2]}")
print(f"  Message: {parts[3]}")
````

````python
# ── F-STRINGS — THE ESSENTIAL WAY TO BUILD TEXT ──────────────────────────
# Prefix the string with f"..." and put variables inside {}
# IT analogy: like variable interpolation in bash ($VAR) or PowerShell ($var)

server   = "prod-db-01"
cpu      = 87.456
memory   = 14.2
status   = "WARNING"

# Basic f-string
message = f"Server {server} | CPU: {cpu}% | Memory: {memory} GB"
print(message)

# Format numbers inside f-strings
print(f"CPU usage  : {cpu:.1f}%")       # 1 decimal place → 87.5%
print(f"CPU usage  : {cpu:.0f}%")       # 0 decimal places → 87%
print(f"Memory     : {memory:06.2f} GB") # pad to width 6 → 014.20 GB

# Multi-line f-string (triple quotes)
alert = f"""
=== SYSTEM ALERT ===
Server  : {server}
Status  : {status}
CPU     : {cpu:.1f}%
Memory  : {memory} GB
===================="""
print(alert)

# Expressions inside f-strings
total = 500
used  = 234.7
print(f"Disk usage: {(used/total)*100:.1f}% ({used}/{total} GB)")
````

````python
# ── USEFUL STRING METHODS FOR IT WORK ────────────────────────────────────

# strip() — removes whitespace from both ends (like trimming in other languages)
raw_input = "   192.168.1.100   "
clean_ip  = raw_input.strip()
print(f"Cleaned IP: '{clean_ip}'")

# join() — combine a list of strings with a separator
servers   = ["web-01", "web-02", "web-03"]
server_list = ", ".join(servers)
print(f"Servers: {server_list}")

# Check if a string contains a substring
log = "2024-01-15 08:11:02 ERROR Database connection refused"
if "ERROR" in log:
    print(f"⚠️  Error detected in log: {log[20:]}")

# Find position of substring
pos = log.find("ERROR")
print(f"'ERROR' found at position: {pos}")

# Count occurrences
sample_log = "ERROR timeout ERROR disk ERROR memory"
error_count = sample_log.count("ERROR")
print(f"Error count: {error_count}")
````

### ✅ Section 2 Checkpoint
- f-strings: `f"Hello {variable}"` — the standard way to embed values in text
- Format numbers: `{value:.2f}` (2 decimals), `{value:.0f}` (integer)
- Key methods: `.split()`, `.strip()`, `.join()`, `.replace()`, `.upper()`, `in` operator

## 📋 Section 3 — Lists, Tuples, Dictionaries & Sets
### ⏱️ Target: 9 minutes

**IT Analogies (one line each):**
> **List** `[]` — ordered inventory you can grow/shrink (mutable, duplicates OK).  
> **Tuple** `()` — fixed snapshot or record (immutable); safe as dict keys.  
> **Dict** `{}` — key–value config / JSON object (mutable, keys unique).  
> **Set** `{x, y}` or `set()` — unique IDs or tags; fast “is this IP in the allowlist?” checks.

````python
# ── LISTS ─────────────────────────────────────────────────────────────────
# Ordered, indexed, allows duplicates, can hold mixed types

servers = ["web-01", "web-02", "db-01", "cache-01"]

# Access by index (starts at 0, like most languages)
print(f"First  : {servers[0]}")    # web-01
print(f"Last   : {servers[-1]}")   # cache-01  ← negative index counts from end
print(f"Count  : {len(servers)}")  # 4

# Slicing — get a subset  [start:end]  (end is exclusive)
web_servers = servers[0:2]          # ['web-01', 'web-02']
print(f"Web servers: {web_servers}")

# Modify a list
servers.append("db-02")             # add to end
servers.insert(0, "lb-01")          # insert at position 0
servers.remove("cache-01")          # remove by value
print(f"\nUpdated list: {servers}")

# Check membership
print(f"'db-01' in list? {('db-01' in servers)}")

# Sort
servers.sort()
print(f"Sorted: {servers}")
````

````python
# ── LIST OF NUMBERS — USEFUL OPERATIONS ──────────────────────────────────

response_times_ms = [120, 340, 89, 456, 210, 78, 1200, 95]

print(f"All response times : {response_times_ms}")
print(f"Count              : {len(response_times_ms)}")
print(f"Minimum            : {min(response_times_ms)} ms")
print(f"Maximum            : {max(response_times_ms)} ms")
print(f"Total              : {sum(response_times_ms)} ms")
print(f"Average            : {sum(response_times_ms)/len(response_times_ms):.1f} ms")

# Filter — find values above threshold
# This is called a 'list comprehension' — a compact way to filter a list
# Read it as: "give me x for each x in the list, IF x > 500"
slow_responses = [x for x in response_times_ms if x > 500]
print(f"\nSlow responses (>500ms): {slow_responses}")
print(f"Count of slow          : {len(slow_responses)}")
````

````python
# ── TUPLES ────────────────────────────────────────────────────────────────
# Ordered like a list, but IMMUTABLE (cannot append/remove after creation).
# IT uses: fixed coordinates, (host, port) pairs, function return bundles.

endpoint = ("10.0.1.42", 443, "HTTPS")   # parentheses optional but recommended
print(f"IP: {endpoint[0]}, port: {endpoint[1]}, scheme: {endpoint[2]}")

# Unpacking — assign names in one line
ip, port, scheme = endpoint
print(f"Unpacked → {ip}:{port} ({scheme})")

# Tuples as dict keys (lists are NOT allowed as keys — they're mutable)
health_by_endpoint = {
    ("10.0.1.1", 8080): "OK",
    ("10.0.1.2", 8080): "DEGRADED",
}
print(health_by_endpoint[("10.0.1.1", 8080)])

# list ↔ tuple when you need to freeze or unfreeze
servers_mutable = ["web-01", "web-02"]
servers_frozen = tuple(servers_mutable)
print(f"Same data, tuple: {servers_frozen}")
````

````python
# ── DICTIONARIES ──────────────────────────────────────────────────────────
# Key-value store. Keys must be unique. Like a JSON object or a config file.

server = {
    "hostname":    "prod-db-01",
    "ip":          "10.0.1.42",
    "cpu_percent": 87.4,
    "memory_gb":   14.2,
    "port":        5432,
    "is_primary":  True,
    "status":      "WARNING"
}

# Access values by key
print(f"Hostname : {server['hostname']}")
print(f"CPU      : {server['cpu_percent']}%")
print(f"Primary  : {server['is_primary']}")

# Safe access with .get() — returns None (or default) if key doesn't exist
# IT analogy: like getenv() with a default value
region  = server.get("region", "us-east-1")   # key missing → use default
print(f"Region   : {region}")

# Add or update a key
server["region"]    = "eu-west-1"
server["cpu_percent"] = 91.2         # update existing

# Useful dict operations
print(f"\nAll keys   : {list(server.keys())}")
print(f"All values : {list(server.values())}")
print(f"Key count  : {len(server)}")
print(f"'port' exists? {'port' in server}")
````

````python
# ── SETS ──────────────────────────────────────────────────────────────────
# Unordered collection of UNIQUE items. Great for tags, allowlists, deduping.

allow_ips = {"10.0.0.1", "10.0.0.2", "10.0.0.5"}
candidate = "10.0.0.2"
print(f"{candidate} allowed? {candidate in allow_ips}")

# Duplicates disappear automatically
log_hosts = ["db-01", "web-01", "db-01", "web-01", "cache-01"]
unique_hosts = set(log_hosts)
print(f"Unique hosts ({len(unique_hosts)}): {sorted(unique_hosts)}")

# Set algebra — compare two groups (subnets, security groups, file IDs…)
team_a = {"ticket-101", "ticket-102", "ticket-103"}
team_b = {"ticket-102", "ticket-104"}
print("Either team touched:", team_a | team_b)           # union
print("Both teams touched:", team_a & team_b)           # intersection
print("Only team A:", team_a - team_b)                  # difference
````

````python
# ── LIST OF DICTS — THE MOST COMMON PATTERN IN REAL CODE ─────────────────
# Like a table: list = rows, dict = columns
# This is exactly what you get back from an API or a database query

fleet = [
    {"hostname": "web-01", "role": "web",   "cpu": 45.0, "status": "OK"},
    {"hostname": "web-02", "role": "web",   "cpu": 67.3, "status": "OK"},
    {"hostname": "db-01",  "role": "db",    "cpu": 91.5, "status": "CRITICAL"},
    {"hostname": "db-02",  "role": "db",    "cpu": 23.1, "status": "OK"},
    {"hostname": "lb-01",  "role": "proxy", "cpu": 12.4, "status": "OK"},
]

print("=" * 45)
print(f"{'Hostname':<10} {'Role':<8} {'CPU%':>6} {'Status'}")
print("=" * 45)
for server in fleet:
    flag = " ⚠️" if server["cpu"] > 80 else ""
    print(f"{server['hostname']:<10} {server['role']:<8} {server['cpu']:>5.1f}% {server['status']}{flag}")
print("=" * 45)

# Filter: find critical servers
critical = [s for s in fleet if s["status"] == "CRITICAL"]
print(f"\n⚠️  Critical servers: {[s['hostname'] for s in critical]}")
````

### List vs Tuple vs Dict vs Set — at a glance

| | **List** `[]` | **Tuple** `()` | **Dict** `{}` | **Set** `{a,b}` / `set()` |
|---|----------------|----------------|---------------|---------------------------|
| **Order** | Yes (indexed) | Yes (indexed) | Insertion order kept (3.7+) | No reliable order |
| **Duplicates** | Allowed | Allowed | Keys must be unique | Elements unique |
| **Mutable?** | Yes | No | Yes | Yes (add/remove elements) |
| **Access** | `items[0]` | `t[0]` | `d["key"]`, `d.get("k", default)` | Membership: `x in s` |
| **Typical IT use** | Growing logs, queues, rows | Fixed records, keys, return pairs | JSON-like config, lookups | Allowlists, dedupe, set ops |

**One-sentence picks:** ordered & changeable → **list**; ordered & fixed → **tuple**; label → value → **dict**; “is it in the group?” / unique only → **set**.

### ✅ Section 3 Checkpoint

| Structure | Best for | Access / ops |
|-----------|----------|----------------|
| **List** `[]` | Ordered, changeable sequences | `items[0]`, slice, `.append()` |
| **Tuple** `()` | Fixed records, dict keys, unpacking | `t[0]`, unpack `a, b = t` |
| **Dict** `{}` | Named fields, JSON-like data | `d['k']`, `.get('k', default)` |
| **Set** `{…}` / `set()` | Unique items, allowlists, dedupe | `x in s`, `\|`, `&`, `-` |
| **List of Dicts** | API/DB rows (table) | `rows[0]['field']` |

## 🔀 Section 4 — Conditions & Loops
### ⏱️ Target: 8 minutes

**IT Analogy:**
> **if/elif/else** = the routing logic in any script.  
> **for loop** = a batch job — "for each server in the inventory, do X".

````python
# ── IF / ELIF / ELSE ──────────────────────────────────────────────────────
# Note: NO parentheses needed around the condition (but they're allowed)
#       NO curly braces — the colon : and indentation define the block

cpu_usage = 87.4

if cpu_usage >= 90:
    print("🔴 CRITICAL — Page the on-call engineer")
elif cpu_usage >= 80:
    print("🟡 WARNING — Send alert to Slack")
elif cpu_usage >= 60:
    print("🟠 ELEVATED — Monitor closely")
else:
    print("🟢 OK — All good")

# Comparison operators
print("\n--- Comparison operators ---")
print(f"5 == 5  → {5 == 5}")    # equal
print(f"5 != 4  → {5 != 4}")    # not equal
print(f"5 > 3   → {5 > 3}")     # greater than
print(f"5 >= 5  → {5 >= 5}")    # greater than or equal

# Logical operators: and, or, not
status    = "WARNING"
disk_full = True

if status == "WARNING" and disk_full:
    print("\n⚠️  Warning + disk full — escalate immediately")

if status == "OK" or disk_full:
    print("At least one concern — continue monitoring")
````

````python
# ── FOR LOOPS ─────────────────────────────────────────────────────────────
# IT analogy: a for loop is a batch job — run the same steps for each item

servers = ["web-01", "web-02", "db-01", "lb-01"]

# Basic for loop — iterate over a list
print("Pinging servers:")
for server in servers:
    print(f"  → Pinging {server}... OK")

# enumerate() — get index AND value (like a for-each with counter)
print("\nNumbered list:")
for index, server in enumerate(servers, start=1):
    print(f"  {index}. {server}")

# range() — loop a specific number of times (like a for i=0; i<n; i++ loop)
print("\nCounting down:")
for i in range(5, 0, -1):   # start=5, stop=0, step=-1
    print(f"  {i}...")
print("  Launch!")
````

````python
# ── LOOPS WITH LOGIC — THE REAL POWER ────────────────────────────────────

fleet = [
    {"hostname": "web-01", "cpu": 45.0, "memory_gb": 6.2},
    {"hostname": "web-02", "cpu": 82.3, "memory_gb": 12.8},
    {"hostname": "db-01",  "cpu": 95.1, "memory_gb": 28.4},
    {"hostname": "db-02",  "cpu": 23.1, "memory_gb": 9.1},
    {"hostname": "lb-01",  "cpu": 12.4, "memory_gb": 3.0},
]

# Collect alerts as we loop
alerts       = []
total_cpu    = 0
server_count = 0

for server in fleet:
    total_cpu    += server["cpu"]   # += means "add to current value"
    server_count += 1

    if server["cpu"] >= 90:
        alerts.append(f"🔴 CRITICAL CPU on {server['hostname']}: {server['cpu']}%")
    elif server["cpu"] >= 80:
        alerts.append(f"🟡 HIGH CPU on {server['hostname']}: {server['cpu']}%")

avg_cpu = total_cpu / server_count

print(f"Fleet summary: {server_count} servers | Avg CPU: {avg_cpu:.1f}%")
print(f"Alerts raised: {len(alerts)}")
for alert in alerts:
    print(f"  {alert}")

# break and continue — control loop flow
print("\n--- break: stop when first critical found ---")
for server in fleet:
    if server["cpu"] >= 90:
        print(f"First critical server: {server['hostname']}")
        break   # stop the loop immediately

print("\n--- continue: skip web servers, process only db ---")
for server in fleet:
    if not server["hostname"].startswith("db"):
        continue   # skip to next iteration
    print(f"DB server: {server['hostname']} | CPU: {server['cpu']}%")
````

````python
# ── WHILE LOOPS — USE WHEN YOU DON'T KNOW THE COUNT ─────────────────────
# IT analogy: like a retry loop — keep trying until success or max attempts

import time   # standard library module for time operations

max_retries = 3
attempt     = 0
connected   = False

while attempt < max_retries and not connected:
    attempt += 1
    print(f"Attempt {attempt}/{max_retries}: connecting to prod-db-01...")
    
    # Simulate: fail on first 2, succeed on 3rd
    if attempt == 3:
        connected = True
        print("  ✅ Connected!")
    else:
        print("  ❌ Connection refused. Retrying in 1 second...")
        time.sleep(1)   # wait 1 second

if not connected:
    print("⛔ Failed to connect after max retries — check firewall rules")
````

### ✅ Section 4 Checkpoint

| Construct | Use for | Key syntax |
|-----------|---------|------------|
| `if/elif/else` | Branching logic | `:` + indent, no `()` required |
| `for x in list` | Iterate over items | Works on lists, dicts, strings |
| `for i in range(n)` | Counted loop | `range(start, stop, step)` |
| `while condition` | Loop until something changes | Retry patterns, polling |
| `break` / `continue` | Control flow inside loops | Same as other languages |

## 🔧 Section 5 — Functions
### ⏱️ Target: 7 minutes

**IT Analogy:**
> A function is a **reusable script module** — define once, call many times with different inputs.  
> Like a PowerShell function or a bash script that takes arguments.

````python
# ── DEFINING AND CALLING FUNCTIONS ───────────────────────────────────────
# def = define.  function_name(parameters):  then indented body
# return sends a value back to the caller

def greet(name):
    """Greet a learner. The triple-quoted string is a docstring — like a comment but discoverable."""
    message = f"Welcome to the AI course, {name}!"
    return message

# Call the function
result = greet("Mayank")
print(result)
print(greet("everyone"))   # call with different input

# ── MULTIPLE PARAMETERS ───────────────────────────────────────────────────
def calculate_disk_usage(used_gb, total_gb):
    """Calculate disk usage percentage and return a status string."""
    percent = (used_gb / total_gb) * 100
    
    if percent >= 90:
        status = "CRITICAL"
    elif percent >= 80:
        status = "WARNING"
    else:
        status = "OK"
    
    return percent, status   # return multiple values as a tuple

pct, status = calculate_disk_usage(234.7, 500)
print(f"\nDisk usage: {pct:.1f}% — Status: {status}")
````

````python
# ── DEFAULT PARAMETER VALUES ──────────────────────────────────────────────
# Parameters can have default values — caller can omit them
# IT analogy: like a function with optional flags

def connect_to_db(host, port=5432, timeout=30, ssl=True):
    """Simulate a database connection with sensible defaults."""
    protocol = "ssl" if ssl else "plain"
    return f"Connecting to {host}:{port} via {protocol} (timeout={timeout}s)"

# Use defaults
print(connect_to_db("prod-db-01"))

# Override specific defaults by name (keyword arguments)
print(connect_to_db("prod-db-01", port=5433, ssl=False))
print(connect_to_db("dev-db-01", timeout=5))

# ── FUNCTIONS THAT PROCESS LISTS ─────────────────────────────────────────
def summarise_fleet(servers):
    """Return a summary dict for a fleet of servers."""
    total      = len(servers)
    avg_cpu    = sum(s["cpu"] for s in servers) / total
    critical   = [s["hostname"] for s in servers if s["cpu"] >= 90]
    return {
        "total_servers": total,
        "avg_cpu":       round(avg_cpu, 1),
        "critical":      critical,
        "healthy":       total - len(critical)
    }

fleet = [
    {"hostname": "web-01", "cpu": 45.0},
    {"hostname": "web-02", "cpu": 82.3},
    {"hostname": "db-01",  "cpu": 95.1},
    {"hostname": "db-02",  "cpu": 23.1},
]

summary = summarise_fleet(fleet)
print(f"\nFleet Summary: {summary}")
print(f"Critical servers: {summary['critical']}")
````

````python
# ── BUILT-IN FUNCTIONS — PYTHON'S STANDARD TOOLKIT ───────────────────────
# These are always available — no import needed

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]

print(f"len()    → {len(numbers)}")         # count
print(f"sum()    → {sum(numbers)}")         # total
print(f"min()    → {min(numbers)}")         # smallest
print(f"max()    → {max(numbers)}")         # largest
print(f"sorted() → {sorted(numbers)}")      # returns sorted copy (doesn't modify original)
print(f"set()    → {set(numbers)}")         # remove duplicates
print(f"list()   → {list(set(numbers))}")   # convert set back to list

# abs(), round()
print(f"\nabs(-42)        → {abs(-42)}")
print(f"round(3.14159, 2) → {round(3.14159, 2)}")

# input() — read from user (useful for interactive scripts)
# name = input("Enter your name: ")   # uncomment to try
print("\n✅ Built-in functions are always available — no imports needed")
````

````python
# ── WRITE A FILE ─────────────────────────────────────────────────────────
# First, let's create a sample log file to read back

sample_log_content = """2024-01-15 08:00:01 INFO  Server started on port 8080
2024-01-15 08:04:12 WARNING  Disk usage at 78%
2024-01-15 08:11:02 ERROR  Database query timeout after 30s
2024-01-15 08:11:05 INFO  Retrying database query (attempt 2/3)
2024-01-15 08:11:08 INFO  Database query succeeded on retry
2024-01-15 08:22:07 ERROR  SMTP timeout connecting to mail server
2024-01-15 08:31:22 WARNING  Memory usage at 71%
2024-01-15 08:33:09 ERROR  Authentication failure: 3 failed attempts
2024-01-15 08:47:33 WARNING  Disk usage at 81% — threshold exceeded
2024-01-15 09:00:00 INFO  Health check: all services nominal
"""

# Write the file
# 'with open()' is the safe way — automatically closes the file
# 'w' = write mode (creates or overwrites)
with open("sample.log", "w") as file:
    file.write(sample_log_content)

print("✅ sample.log created")
````

````python
# ── READ A FILE ───────────────────────────────────────────────────────────
# 'r' = read mode (default)

with open("sample.log", "r") as file:
    lines = file.readlines()   # read all lines into a list

print(f"Lines in file: {len(lines)}")
print(f"First line   : {lines[0].strip()}")
print(f"Last line    : {lines[-1].strip()}")

# Count log levels
error_count   = sum(1 for line in lines if " ERROR " in line)
warning_count = sum(1 for line in lines if " WARNING " in line)
info_count    = sum(1 for line in lines if " INFO " in line)

print(f"\nLog summary:")
print(f"  ERRORs   : {error_count}")
print(f"  WARNINGs : {warning_count}")
print(f"  INFOs    : {info_count}")
````

````python
# ── ERROR HANDLING WITH TRY / EXCEPT ─────────────────────────────────────
# IT analogy: like try-catch in Java/C# — handle failures without crashing

# Example 1: File not found
def read_log_safe(filename):
    """Read a log file safely — returns empty list if file doesn't exist."""
    try:
        with open(filename, "r") as f:
            return f.readlines()
    except FileNotFoundError:
        print(f"⚠️  File not found: {filename}")
        return []
    except PermissionError:
        print(f"⚠️  Permission denied: {filename}")
        return []

lines = read_log_safe("sample.log")
print(f"Read {len(lines)} lines from sample.log")

lines_missing = read_log_safe("does_not_exist.log")
print(f"Read {len(lines_missing)} lines from missing file")

# Example 2: Type conversion safely
def safe_int(value, default=0):
    """Convert to int, return default if conversion fails."""
    try:
        return int(value)
    except (ValueError, TypeError):
        return default

print(f"\nsafe_int('42')    → {safe_int('42')}")
print(f"safe_int('abc')   → {safe_int('abc')}")
print(f"safe_int('abc',9) → {safe_int('abc', 9)}")
print(f"safe_int(None)    → {safe_int(None)}")
````

## 🏗️ Section 7 — Mini-Project: Log File Analyser
### ⏱️ Target: 5 minutes

**Put it all together.** This uses every concept from this session:
variables, strings, lists, dicts, loops, conditions, functions, file I/O.

> **This is the exact same pattern you'll use in Week 2** — except instead of printing the summary, you'll send it to an AI API and get an intelligent analysis back.

````python
# ── GENERATE sample.log (run this before the analyser) ─────────────────────
# Creates sample.log in the notebook's working directory

sample_log_content = """2024-01-15 08:00:01 INFO  Server started on port 8080
2024-01-15 08:04:12 WARNING  Disk usage at 78%
2024-01-15 08:11:02 ERROR  Database query timeout after 30s
2024-01-15 08:11:05 INFO  Retrying database query (attempt 2/3)
2024-01-15 08:11:08 INFO  Database query succeeded on retry
2024-01-15 08:22:07 ERROR  SMTP timeout connecting to mail server
2024-01-15 08:31:22 WARNING  Memory usage at 71%
2024-01-15 08:33:09 ERROR  Authentication failure: 3 failed attempts
2024-01-15 08:47:33 WARNING  Disk usage at 81% — threshold exceeded
2024-01-15 09:00:00 INFO  Health check: all services nominal
"""

with open("sample.log", "w") as file:
    file.write(sample_log_content)

print("✅ sample.log created")
````

````python
# ── MINI-PROJECT: LOG FILE ANALYSER ──────────────────────────────────────
# Reads a log file and produces a structured summary report
# Uses: variables, f-strings, lists, dicts, loops, conditions, functions, files

def parse_log_line(line):
    """
    Parse one log line into a structured dict.
    Input  : '2024-01-15 08:11:02 ERROR  Database query timeout'
    Output : {'date': '2024-01-15', 'time': '...', 'level': 'ERROR', 'message': '...'}
    """
    line = line.strip()
    if not line:
        return None
    
    parts = line.split(" ", 3)   # split into max 4 parts
    if len(parts) < 4:
        return None
    
    return {
        "date":    parts[0],
        "time":    parts[1],
        "level":   parts[2].strip(),
        "message": parts[3]
    }


def analyse_log_file(filename):
    """
    Read a log file and return a full analysis summary.
    """
    # Read the file
    try:
        with open(filename, "r") as f:
            raw_lines = f.readlines()
    except FileNotFoundError:
        return {"error": f"File not found: {filename}"}

    # Parse each line
    parsed      = []
    errors      = []
    warnings    = []
    counts      = {"ERROR": 0, "WARNING": 0, "INFO": 0, "OTHER": 0}

    for line in raw_lines:
        entry = parse_log_line(line)
        if not entry:
            continue
        
        parsed.append(entry)
        level = entry["level"]

        if level in counts:
            counts[level] += 1
        else:
            counts["OTHER"] += 1

        if level == "ERROR":
            errors.append(entry)
        elif level == "WARNING":
            warnings.append(entry)

    # Build summary
    return {
        "filename":     filename,
        "total_lines":  len(parsed),
        "counts":       counts,
        "errors":       errors,
        "warnings":     warnings,
        "first_entry":  parsed[0]  if parsed else None,
        "last_entry":   parsed[-1] if parsed else None,
    }


def print_report(summary):
    """
    Print a formatted report from the analysis summary dict.
    """
    if "error" in summary:
        print(f"❌ {summary['error']}")
        return

    c = summary["counts"]
    print("\n" + "=" * 55)
    print(f"  LOG ANALYSIS REPORT — {summary['filename']}")
    print("=" * 55)
    print(f"  Total lines   : {summary['total_lines']}")
    print(f"  Period        : {summary['first_entry']['date']} {summary['first_entry']['time']}"
          f" → {summary['last_entry']['date']} {summary['last_entry']['time']}")
    print()
    print(f"  {'Level':<10} {'Count':>5}  {'Bar'}")
    print(f"  {'-'*38}")
    for level, count in c.items():
        if count == 0:
            continue
        bar   = "█" * count
        emoji = {"ERROR":"🔴", "WARNING":"🟡", "INFO":"🟢", "OTHER":"⚪"}.get(level, "")
        print(f"  {emoji} {level:<8} {count:>5}  {bar}")

    if summary["errors"]:
        print(f"\n  ── ERRORS ({len(summary['errors'])}) ──")
        for e in summary["errors"]:
            print(f"  [{e['time']}] {e['message']}")

    if summary["warnings"]:
        print(f"\n  ── WARNINGS ({len(summary['warnings'])}) ──")
        for w in summary["warnings"]:
            print(f"  [{w['time']}] {w['message']}")

    print("=" * 55)


# ── RUN THE ANALYSER ──────────────────────────────────────────────────────
summary = analyse_log_file("sample.log")
print_report(summary)
````
